```python
 #for ordinal data(label-encode)
#manually use dict if you want the number to rep order
clarity_dict = {'IF' : 7, 'VVS1' : 6, 'VVS2' : 5, 'VS1' : 4, 'VS2' : 3, 'SI1' : 2, 'SI2' : 1, 'I1' : 0}

data['clarity'].replace(clarity_dict, inplace=True)

data.head()
```

```python
#one-hot-encoding
data = pd.get_dummies(data, columns=['cut', 'color'])
data.head()
```

```python
NUMERIC_FEATURES = ['carat', 'depth', 'table', 'x', 'y', 'z']
numeric_df =data[NUMERIC_FEATURES]

numeric_df.describe()
```

```python
#scaling
from sklearn.preprocessing import scale

numeric_array = scale(numeric_df)

numeric_array[:5]
```

```python
#convert to dataframe
numeric_df = pd.DataFrame(numeric_array, columns=numeric_df.columns)
numeric_df.describe()
```

```python
#non standardized
data.head()
```

```python
#replace with standardized ones
data.drop(NUMERIC_FEATURES, inplace=True, axis=1)
data.head()
```

```python
#concatinate
data = pd.concat([data, numeric_df], axis=1)
data.shape
```

```python
data.head()
```

```python
data.to_csv(r'C:\Users\clozy\opencv\Scripts\jpnb\CREATING\Regression\2.0\diamonds_processed.csv')
```

```python
from sklearn.model_selection import train_test_split

X = data.drop('price', axis=1)
Y = data['price']

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

linear_regression = LinearRegression()
linear_regression.fit(x_train, y_train)
```

```python
y_pred = linear_regression.predict(x_test)

y_pred
```

```python
train_score = linear_regression.score(x_train, y_train)
print('training score : ', train_score)
```

```python
from sklearn.metrics import r2_score

score = r2_score(y_test, y_pred)

print('Testing score : ', score )
```
